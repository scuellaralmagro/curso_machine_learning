### Definir una clase

Para crear un objeto primero debemos definir la clase que lo contiene. Podemos pensar en la *clase* como el molde con el que crear nuevos objetos de ese tipo:

queque.png

> Usaremos la palabra reservada `class` para crear nuevas clases.

```python
class Persona():
    
    # La función __init__ está predeterminada en python y se ejecuta una vez cuando creamos la clase.
    # Todas las funciones deben incluir la clase a la que pertenecen dentro de sus parametros: self.
    
    def __init__(self):
        
        # Las variables dentro de una clase se indican con el prefijo self
        self.nombre = "Juan"
        self.apellido = "Sanchís"
        
    def presentarse(self):
        
        # Cuando ejecutemos la siguiente función la clase Persona() imprimirá un saludo
        print("¡Hola!, me llamo", self.nombre, self.apellido)
```

```python
# Creamos una variable de tipo Persona
persona = Persona()
```

```python
# Ejecutamos la función de presentarse
persona.presentarse()
```

## 🔦 Usos de doble subguión `__`

Los nombres que comienzan y terminan con dos subguiones `__` están reservados para uso interno de Python, así que no se deberían utilizar en código propio. Estos nombres se conocen como **dunder** que proviene de "*double-underscore*".

Veamos un ejemplo en el que se muestra el nombre de una función y su documentación:

```python
def amazing():
    '''This is the amazing function.
    Want to see it again?'''
    print('This function is named:', amazing.__name__)
    print('And its docstring is:', amazing.__doc__)
```

```python
amazing()
```

> El programa principal se asigna a una variable especial llamada `__main__`.

```python
# Hemos creado la clase Circulo
import matplotlib.pyplot as plt

class Circulo(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Method
    def anyadir_radio(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def dibujarCirculo(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
```

Si la palabra dada como argumento no aparece en el texto, devuelve 0. Puedes verificar si un string es una clave en el diccionario usando la siguiente fórmula de código: <code>if item in my_dictionary:</code>

```python
class analysedText(object):
    
    def __init__ (self, text):
        lowered = text.lower()
        no_point = lowered.replace('.', '')
        no_exclamation = no_point.replace('!', '')
        no_comma = no_exclamation.replace(',', '')
        no_question = no_comma.replace('?', '')

        self.fmtText = no_question
    
    def freqAll(self):    
        fmt_text_list = self.fmtText.split(' ')

        fmt_text_set = set(fmt_text_list)
        fmt_text_count = []
        for word in fmt_text_set:
            fmt_text_count.append(fmt_text_list.count(word))
        freqAll_dict = dict(zip(fmt_text_set, fmt_text_count))
        return freqAll_dict
    
    def freqOf(self, word):
        fmt_text_list = self.fmtText.split(' ')
        freqOf_count = fmt_text_list.count(word)
        return freqOf_count
        
```
